```python
import pandas as pd

df = pd.read_csv('Loblaws_csv.csv')

print(df) 
```

```python
# Display data type

df.dtypes
```

```python
# Dropping duplicates and sorting by 'product name'

df_final = df.drop_duplicates().sort_values(by='product name', ascending=False)
df_final
```

```python
df_final.to_csv('Loblaw_final.csv')
```

```python
# Sort by product brand

filtered_df = df_final[df_final['product brand'] == 'Unknown']
filtered_df
```

```python
filtered_df.to_csv('Loblaw_final_brand.csv')
```
