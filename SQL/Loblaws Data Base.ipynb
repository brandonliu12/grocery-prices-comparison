```python
import pandas as pd
```

```python
#Read CSV File
file = 'Unique_Loblaws_price_comparison.CSV'
loblaws_df = pd.read_csv(file)
loblaws_df['brand id'] = pd.factorize(loblaws_df['product brand'])[0] + 1
loblaws_df['store id'] = pd.factorize(loblaws_df['store name'])[0] + 1
loblaws_df
```

```python
zucchini_df = loblaws_df[loblaws_df['product name'] == '100% Carrot']
zucchini_df.head()
```

```python
#CREATE STORE CSV
store_df = pd.DataFrame(loblaws_df['store name'].unique(), columns=['store name'])
store_df['store id'] = pd.factorize(store_df['store name'])[0] + 1
store_df = store_df[['store id', 'store name']]
store_df.to_csv('loblaws_stores.csv', index = False)
store_df.head()
```

```python
#CREATE BRAND CSV
brand_df = pd.DataFrame(loblaws_df['product brand'].unique(), columns=['product brand'])
brand_df['brand id'] = pd.factorize(brand_df['product brand'])[0] + 1
brand_df = brand_df[['brand id', 'product brand']]
brand_df.to_csv('loblaws_brands.csv', index = False)
brand_df.head()
```

```python
#CREATE EDMONTON CSV
# Filter the Edmonton-specific data
edmonton_df = loblaws_df[loblaws_df['store name'] == 'Brewery Dist, Edmonton'].copy()

# Rename the 'price' column to 'edmonton_price'
edmonton_df.rename(columns={'price': 'edmonton_price'}, inplace=True)

# Drop the 'store name' column
edmonton_df = edmonton_df.drop('store name', axis=1)
edmonton_df = edmonton_df.drop('store id', axis=1)

# Save the DataFrame to a CSV file
edmonton_df.to_csv('Loblaws_Edmonton.csv', index=False)
```

```python
edmonton_df.head()
```

```python
#CREATE TORONTO CSV
# Filter the Toronto-specific data
toronto_df = loblaws_df[loblaws_df['store name'] == 'Mclaughlin, Toronto'].copy()

# Rename the 'price' column to 'toronto_price'
toronto_df.rename(columns={'price': 'toronto_price'}, inplace=True)

# Drop the 'store name' column
toronto_df = toronto_df.drop('store name', axis=1)
toronto_df = toronto_df.drop('store id', axis=1)

# Save the DataFrame to a CSV file
toronto_df.to_csv('Loblaws_Toronto.csv', index=False)
```
