```python
# Import Splinter and BeautifulSoup
from splinter import Browser
from bs4 import BeautifulSoup as soup
import re

executable_path = r'C:\Users\liu12\Downloads\chromedriver_win32\chromedriver.exe'
browser = Browser('chrome', executable_path=executable_path)

# Visit loblaws
url = 'https://www.walmart.ca/browse/grocery/fruits-vegetables/fresh-fruits/10019-6000194327370-6000194327411?icid=browse_l2_grocery_fresh_fruits_961_CKZLUB76QL&fromFC=true'
browser.visit(url)

# Create a Beautiful Soup object
html = browser.html
walmart_soup = soup(html, 'html.parser')
print(walmart_soup.prettify())
```

```python
# Find all grocery and their prices
produce = walmart_soup.find_all('a', class_="css-mfluef e1m8uw911")
grocery_list = []

# Create a for loop to extract the information
for grocery in produce:
    product_title = grocery.find("p", class_="css-1p4va6y eudvd6x0")
    product = product_title.text if product_title else "Unknown"
    print(f"Product Title: {product}")  # Debug print
    
    price_item = grocery.find("span", class_="css-2vqe5n esdkp3p0")
    price = float(re.search(r'[\d.]+', price_item.text).group()) if price_item else 0.0
    print(f"Price: {price}")  # Debug print
    
    grocery_dict = {'product brand': product, "price": price}
    grocery_list.append(grocery_dict)
```
